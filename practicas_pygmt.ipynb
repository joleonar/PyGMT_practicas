```python
import pygmt
import pandas as pd
```

```python
data = pd.read_csv("datos.txt", sep='\t') 
data.head()
```

```python
fig = pygmt.Figure()
fig.coast(
    region=[-74,-59,5,14],
    projection='M8i',
    shorelines=True,
    borders=['1/0.8p','2/0.1p'],
    water='skyblue',
    land='lightyellow',
    frame=True,
    resolution='f'
)

fig.plot(x=data.LON, y=data.LAT, style="c0.3c", color="white", pen="black")
#fig.plot(x=-62.5, y=10.2, style=)
fig.basemap(frame='a2f1g2')
fig.show()
#fig.show(method='external')
```
